# Establishing nutritional profiles

With the information extracted from the tweets we can determine the differences there are in nutrition in different regions of spain, such as which region consumes the most or the least carbohydrates or proteins. To do so we perform t-tests over our dataset.

```python
df = pd.read_csv('SpainDatasetFood.csv')
df
```

```python
southCoast = df[(df.ComunidadAutonoma == 'Andaluc√≠a') | (df.ComunidadAutonoma == 'Ceuta') | (df.ComunidadAutonoma == 'Melilla') | (df.ComunidadAutonoma == 'Murcia') | (df.ComunidadAutonoma == 'Islas Canarias')]
eastCoast = df[(df.ComunidadAutonoma == 'Catalu√±a') | (df.ComunidadAutonoma == 'Comunidad Valenciana') | (df.ComunidadAutonoma == 'Islas Baleares') | (df.ComunidadAutonoma == 'Arag√≥n')]
northCoast = df[(df.ComunidadAutonoma == 'Galicia') | (df.ComunidadAutonoma == 'Asturias') | (df.ComunidadAutonoma == 'Cantabria') | (df.ComunidadAutonoma == 'Pais Vasco') | (df.ComunidadAutonoma == 'Navarra')]
innerRing = df[(df.ComunidadAutonoma == 'Castilla la Mancha') | (df.ComunidadAutonoma == 'Castilla y Le√≥n') | (df.ComunidadAutonoma == 'Extremadura') | (df.ComunidadAutonoma == 'La Rioja') ]
center = df[df.ComunidadAutonoma == 'Madrid']

"""southCoast = southCoast.mean()
eastCoast = eastCoast.mean()
northCoast = northCoast.mean()
innerRing = innerRing.mean()
center = center.mean()
"""
print(southCoast.mean())
print(eastCoast.mean())
print(northCoast.mean())
print(innerRing.mean())
print(center.mean())
```

```python
from scipy import stats

ceCal = stats.ttest_ind(center['Calorias'],eastCoast['Calorias'])
cnCal = stats.ttest_ind(center['Calorias'],northCoast['Calorias'])
csCal = stats.ttest_ind(center['Calorias'],southCoast['Calorias'])
ciCal = stats.ttest_ind(center['Calorias'],innerRing['Calorias'])
snCal = stats.ttest_ind(southCoast['Calorias'],northCoast['Calorias'])
seCal = stats.ttest_ind(southCoast['Calorias'],eastCoast['Calorias'])
siCal = stats.ttest_ind(southCoast['Calorias'],innerRing['Calorias'])
eiCal = stats.ttest_ind(eastCoast['Calorias'],innerRing['Calorias'])
enCal = stats.ttest_ind(eastCoast['Calorias'],northCoast['Calorias'])
niCal = stats.ttest_ind(northCoast['Calorias'],innerRing['Calorias'])

ceHC = stats.ttest_ind(center['Hidratos de Carbono'],eastCoast['Hidratos de Carbono'])
cnHC = stats.ttest_ind(center['Hidratos de Carbono'],northCoast['Hidratos de Carbono'])
csHC = stats.ttest_ind(center['Hidratos de Carbono'],southCoast['Hidratos de Carbono'])
ciHC = stats.ttest_ind(center['Hidratos de Carbono'],innerRing['Hidratos de Carbono'])
snHC = stats.ttest_ind(southCoast['Hidratos de Carbono'],northCoast['Hidratos de Carbono'])
seHC = stats.ttest_ind(southCoast['Hidratos de Carbono'],eastCoast['Hidratos de Carbono'])
siHC = stats.ttest_ind(southCoast['Hidratos de Carbono'],innerRing['Hidratos de Carbono'])
eiHC = stats.ttest_ind(eastCoast['Hidratos de Carbono'],innerRing['Hidratos de Carbono'])
enHC = stats.ttest_ind(eastCoast['Hidratos de Carbono'],northCoast['Hidratos de Carbono'])
niHC = stats.ttest_ind(northCoast['Hidratos de Carbono'],innerRing['Hidratos de Carbono'])

ceG = stats.ttest_ind(center['Grasas'],eastCoast['Grasas'])
cnG = stats.ttest_ind(center['Grasas'],northCoast['Grasas'])
csG = stats.ttest_ind(center['Grasas'],southCoast['Grasas'])
ciG = stats.ttest_ind(center['Grasas'],innerRing['Grasas'])
snG = stats.ttest_ind(southCoast['Grasas'],northCoast['Grasas'])
seG = stats.ttest_ind(southCoast['Grasas'],eastCoast['Grasas'])
siG = stats.ttest_ind(southCoast['Grasas'],innerRing['Grasas'])
eiG = stats.ttest_ind(eastCoast['Grasas'],innerRing['Grasas'])
enG = stats.ttest_ind(eastCoast['Grasas'],northCoast['Grasas'])
niG = stats.ttest_ind(northCoast['Grasas'],innerRing['Grasas'])

ceP = stats.ttest_ind(center['Proteinas'],eastCoast['Proteinas'])
cnP = stats.ttest_ind(center['Proteinas'],northCoast['Proteinas'])
csP = stats.ttest_ind(center['Proteinas'],southCoast['Proteinas'])
ciP = stats.ttest_ind(center['Proteinas'],innerRing['Proteinas'])
snP = stats.ttest_ind(southCoast['Proteinas'],northCoast['Proteinas'])
seP = stats.ttest_ind(southCoast['Proteinas'],eastCoast['Proteinas'])
siP = stats.ttest_ind(southCoast['Proteinas'],innerRing['Proteinas'])
eiP = stats.ttest_ind(eastCoast['Proteinas'],innerRing['Proteinas'])
enP = stats.ttest_ind(eastCoast['Proteinas'],northCoast['Proteinas'])
niP = stats.ttest_ind(northCoast['Proteinas'],innerRing['Proteinas'])

ceCol = stats.ttest_ind(center['Colesterol'],eastCoast['Colesterol'])
cnCol = stats.ttest_ind(center['Colesterol'],northCoast['Colesterol'])
csCol = stats.ttest_ind(center['Colesterol'],southCoast['Colesterol'])
ciCol = stats.ttest_ind(center['Colesterol'],innerRing['Colesterol'])
snCol = stats.ttest_ind(southCoast['Colesterol'],northCoast['Colesterol'])
seCol = stats.ttest_ind(southCoast['Colesterol'],eastCoast['Colesterol'])
siCol = stats.ttest_ind(southCoast['Colesterol'],innerRing['Colesterol'])
eiCol = stats.ttest_ind(eastCoast['Colesterol'],innerRing['Colesterol'])
enCol = stats.ttest_ind(eastCoast['Colesterol'],northCoast['Colesterol'])
niCol = stats.ttest_ind(northCoast['Colesterol'],innerRing['Colesterol'])

ceF = stats.ttest_ind(center['Fibra'],eastCoast['Fibra'])
cnF = stats.ttest_ind(center['Fibra'],northCoast['Fibra'])
csF = stats.ttest_ind(center['Fibra'],southCoast['Fibra'])
ciF = stats.ttest_ind(center['Fibra'],innerRing['Fibra'])
snF = stats.ttest_ind(southCoast['Fibra'],northCoast['Fibra'])
seF = stats.ttest_ind(southCoast['Fibra'],eastCoast['Fibra'])
siF = stats.ttest_ind(southCoast['Fibra'],innerRing['Fibra'])
eiF = stats.ttest_ind(eastCoast['Fibra'],innerRing['Fibra'])
enF = stats.ttest_ind(eastCoast['Fibra'],northCoast['Fibra'])
niF = stats.ttest_ind(northCoast['Fibra'],innerRing['Fibra'])

compar = ['regi√≥n centro y costa este', 'regi√≥n centro y costa norte', 'regi√≥n centro y costa sur', 'regi√≥n centro y anillo interior', 'costa sur y costa este', 'costa sur y costa norte', 'costa sur y anillo interior', 'costa este y anillo interior', 'costa este y costa norte', 'costa norte y anillo interior']
cal = [ceCal,cnCal,csCal,ciCal,snCal,seCal,siCal,eiCal,enCal,niCal]
hid = [ceHC,cnHC,csHC,ciHC,snHC,seHC,siHC,eiHC,enHC,niHC]
gra = [ceG,cnG,csG,ciG,snG,seG,siG,eiG,enG,niG]
pro = [ceP,cnP,csP,ciP,snP,seP,siP,eiP,enP,niP]
col = [ceCol,cnCol,csCol,ciCol,snCol,seCol,siCol,eiCol,enCol,niCol]
fib = [ceF,cnF,csF,ciF,snF,seF,siF,eiF,enF,niF]

dfttest = pd.DataFrame({'Regi√≥n':compar,'tTest Calorias':cal,'tTest Hidratos de Carbono':hid,'tTest Grasas':gra,'tTest Proteinas':pro,'tTest Colesterol':col,'tTest Fibra':fib})
dfttest
```
